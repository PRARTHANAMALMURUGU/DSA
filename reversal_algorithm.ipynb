```python
def reversal(arr,d):
    n=len(arr)
    d%=n
    arr.reverse()
    arr[:d]=reversed(arr[:d])
    arr[d:]=reversed(arr[d:])
nums=[1,2,3,4,5,6]
reversal(nums,2)
for x in nums:
    print(x,end=" ")
```
