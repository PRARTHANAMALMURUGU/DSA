```python
def freq_count(freq_table, key):
    if key in freq_table:
        freq_table[key] += 1
    else:
        freq_table[key] = 1

freq_table = {}
n = int(input("Enter number of elements you want to add in the list: "))

for _ in range(n):
    x = int(input("Enter the number: "))
    freq_count(freq_table, x)

for key, count in freq_table.items():
    print("Key:", key, "Frequency:", count)
3
```
