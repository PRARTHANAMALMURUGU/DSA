```python
from math import gcd
def juggling(arr,d):
    n=len(arr)
    d%=n
    cycles=gcd(d,n)
    for i in range(cycles):
        curr_index=i
        curr_element=arr[curr_index]
    while True:
        next_index=(curr_index+d)%n
        next_element=arr[next_index]
        arr[next_index]=curr_element
        curr_index=next_index
        curr_element=next_element
        if curr_index==i:
            break
nums=[1,2,3,4,5]
juggling(nums,2)
for x in nums:
    print(x,end=" ")
```
