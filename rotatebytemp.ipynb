```python
def rotate(arr,d):
    n=len(arr)
    temp=[0]*n
    for i in range(d):
        temp[i]=arr[n-d+i]
    for i in range(n-d):
        temp[d+i]=arr[i]
    for i in range(n):
        arr[i]=temp[i]
nums=[1,2,3,4,5,6]
rotate(nums,2)
for x in nums:
    print(x,end=" ")
```
