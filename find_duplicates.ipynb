```python
def duplicates(arr):
  freq={}
  res=[]
  for nums in arr:
    freq[nums]=freq.get(nums,0)+1
    for key,value in freq.items():
      if value>1:
        res.append(key)
  return res
n=[1,2,34,3,2,1,2]
print(set(duplicates(n)))
```
